# Extraer los conceptos claves de un corpus utilizando un enfoque basado en reglas

Este notebook muestra:

- cómo extraer las conceptos claves de un corpus.

Los conceptos claves son construcciones terminologícas como por ejemplo:  "calentamiento global", "crisis climática", "energía renovable", "huella de carbono"



## 1 Cargar el corpus de noticias

```python
import pandas
archivo = "./data/chile_2021-02.csv"
corpus = pandas.read_csv(archivo)
corpus.drop('year', inplace=True, axis=1)
corpus.drop('id_journalist', inplace=True, axis=1)
```

## 4.2 Extracción de conceptos claves en una noticia

```python
N=908
url = corpus.iloc[N]["url"]
url
```

```python
N=908
text = corpus.iloc[N]["text"]
text
```

```python
import spacy
from spacy.matcher import Matcher
nlp = spacy.load("es_core_news_sm")
```

```python
matcher = Matcher(nlp.vocab)
```

- Pattern 1: "NOUN" de "NOUN"

```python
pattern_1 = [{"POS": "NOUN"},{"LOWER": "de"}, {"POS": "NOUN"}]
matcher.add("NOUN-de-NOUN", [pattern_1])
```

- Pattern 2: "NOUN" "ADJ"

```python
pattern_2 = [{"POS": "NOUN"}, {"POS": "ADJ"}]
matcher.add("NOUN-ADJ", [pattern_2])
```

```python
doc = nlp(text)

matches = matcher(doc)

for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
```

## 4.3 Extracción de los conceptos claves en todas las noticias y calculo de la frecuencia

```python
from collections import Counter
```

```python
concept_freq_total=Counter({})

for index, row in corpus.iterrows():
    print(index)
    text = str(row["text"])
    if(len(text)>0):
        doc = nlp(text)
        
        matches = matcher(doc)
        
        for match_id, start, end in matches:
            span = doc[start:end]
            concept = [span.text]
            
            concept_freq = Counter(concept)
            concept_freq_total = concept_freq_total + concept_freq
```

```python
pandas.set_option('display.max_rows', None)

top_n = 2000
data = concept_freq_total.most_common(top_n)
result = pandas.DataFrame(data, columns =['Concept', 'Frequency'])
result
```

```python
result.to_csv("./output/terminology_02_2021.csv")
```
